```python
def polynomial_sframe(feature, degree):
    poly_sframe = graphlab.SFrame()
    poly_sframe["power_1"] = feature
    
    if degree > 1:
        for power in range(2, degree + 1):
            name = "power_" + str(power)
            poly_sframe[name] = feature ** power
            
    
    return poly_sframe
```
